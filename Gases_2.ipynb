**Velocidad vs Número de moléculas**

```python
from drawdata import draw_histogram
draw_histogram()
```

```python
import csv
import matplotlib.pyplot as plt
velocidad = pd.read_csv("velocidad.csv")
df = pd.DataFrame(velocidad)

entrada = open('velocidad.csv')
lector = csv.reader(entrada)

next(lector)

x = []
y = []

for fila in lector:
    x.append(float(fila[0]))
    y.append(float(fila[1]))
factor = 2565 # s
x = [valor * factor for valor in x]


sns.lineplot(
    data = df,
    x= x, y= y, units="z",
    estimator=None, lw=1,)

plt.xlabel('Velocidad molecular (m/s)')
plt.ylabel('Numero de moleculas (n)')
plt.title('Velocidad vs Número de moléculas')



# Muestra el gráfico
plt.show()
```
